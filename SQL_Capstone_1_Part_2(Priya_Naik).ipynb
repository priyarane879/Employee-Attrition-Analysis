#### Find Attrition by Age Groups

```python
pd.read_sql_query('''SELECT Attrition, CASE
         WHEN Age < 30 THEN 'Under 30'
         WHEN Age BETWEEN 30 AND 39 THEN '30-40'
         WHEN Age BETWEEN 40 AND 49 THEN '40-50'
         ELSE 'Over 50'
         END AS Age_group,
         COUNT(*) as num,
         ROUND((COUNT(*) * 100.0 / SUM(COUNT(*))
         OVER (PARTITION  by CASE
                              WHEN Age < 30 THEN 'Under 30'
                              WHEN Age BETWEEN 30 AND 39 THEN '30-40'
                              WHEN Age BETWEEN 40 AND 49 THEN '40-50'
                              ELSE 'Over 50'
                              END )),2) as percent_by_age
         FROM table_HR GROUP BY Attrition, CASE
                                         WHEN Age < 30 THEN 'Under 30'
                                         WHEN Age BETWEEN 30 AND 39 THEN '30-40'
                                         WHEN Age BETWEEN 40 AND 49 THEN '40-50'
                                         ELSE 'Over 50'
                                         END order by percent_by_age desc''',sqliteConnection)
```

Observation:-
1. Younger employees (Under 30) are the most likely to leave, possibly due to career exploration, job dissatisfaction, or better opportunities.

2. Middle-aged employees (30-50) are more stable, likely due to career growth, financial commitments, or job satisfaction.

3. Older employees (Over 50) have low attrition, possibly due to job security, nearing retirement, or fewer external job opportunities.

#### Find Attrition by Years At Company

```python
### Implement code
pd.read_sql_query('''SELECT
    CASE
        WHEN YearsAtCompany < 2 THEN 'New Hires'
        WHEN YearsAtCompany BETWEEN 2 AND 5 THEN '2-5 years'
        WHEN YearsAtCompany BETWEEN 6 AND 10 THEN '6-10 years'
        WHEN YearsAtCompany BETWEEN 11 AND 20 THEN '11-20 years'
        ELSE 'Over 20 years'
    END AS tenure_years,
    COUNT(*) AS num,
    ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(), 1) AS percent
FROM table_HR where Attrition = 'Yes'
GROUP BY
    CASE
        WHEN YearsAtCompany < 2 THEN 'New Hires'
        WHEN YearsAtCompany BETWEEN 2 AND 5 THEN '2-5 years'
        WHEN YearsAtCompany BETWEEN 6 AND 10 THEN '6-10 years'
        WHEN YearsAtCompany BETWEEN 11 AND 20 THEN '11-20 years'
        ELSE 'Over 20 years'
    END
ORDER BY num DESC''',sqliteConnection)
```

Observation:-
1. Early-career employees (New Hires & 2-5 years) are the most likely to leave, indicating that retention strategies should focus on engagement, career growth, and satisfaction within the first 5 years.

2. Mid-career employees (6-10 years) still experience some attrition, but the rate declines significantly, suggesting they find more stability.

3. Long-tenured employees (11+ years) have very low attrition, meaning that those who stay beyond a decade are likely to remain long-term.

Q.2 Why do people with higher pay still leave the company?

```python
pd.read_sql_query('''
       WITH IncomePercentiles AS (
    SELECT
        MonthlyIncome,
        NTILE(4) OVER (ORDER BY MonthlyIncome) AS IncomeGroup
    FROM table_HR
    WHERE MonthlyIncome > (SELECT AVG(MonthlyIncome) FROM table_HR)
)
SELECT
    CASE
        WHEN ip.IncomeGroup = 1 THEN 'Low (0-25th Percentile)'
        WHEN ip.IncomeGroup = 2 THEN 'Medium (25th-50th Percentile)'
        WHEN ip.IncomeGroup = 3 THEN 'High (50th-75th Percentile)'
        WHEN ip.IncomeGroup = 4 THEN 'Very High (75th-100th Percentile)'
    END AS IncomeCategory,
    t.Attrition,
    Avg(t.MonthlyIncome),
    Avg(t.JobSatisfaction),
    Avg(t.WorkLifeBalance),
    Avg(t.YearsSinceLastPromotion),
    Avg(t.PerformanceRating)

FROM table_HR t
JOIN IncomePercentiles ip ON t.MonthlyIncome = ip.MonthlyIncome
WHERE t.MonthlyIncome > (SELECT AVG(MonthlyIncome) FROM table_HR)
GROUP BY IncomeCategory, t.Attrition;
                    ''', sqliteConnection)
```

Observation:-
1. Employees who did not leave have slightly higher average salaries than those who left.
2. Employees with lower job satisfaction are more likely to leave, regardless of salary.
3. Poor Work-Life Balance contributes to attrition, especially in High & Low percentiles.
4. Long promotion gaps strongly correlate with attrition.Even high salaries do not retain employees if promotions are delayed.
5.  Employees with similar performance ratings still leave, indicating attrition is driven more by promotion delays & job satisfaction than performance.

Q.3) Which factors drive employees who work at company less than 5 years to leave?

```python
pd.read_sql_query('''
SELECT
    YearsAtCompany,Attrition,
    COUNT(*) AS Number_of_Employees,
    AVG(MonthlyIncome) AS Average_Monthly_Income,
    AVG(JobLevel) AS Average_JobLevel,
    AVG(JobSatisfaction) AS Average_JobSatisfaction,
    AVG(WorkLifeBalance) AS Average_WorkLifeBalance
FROM table_HR
WHERE YearsAtCompany < 5
GROUP BY YearsAtCompany,Attrition
ORDER BY Attrition
''', sqliteConnection)
```

Observation:-
1. Employees who left had significantly lower average monthly income and job levels than those who stayed.
2. Employees who left had lower work-life balance scores.
3. New hires with lower salaries are more likely to leave early.
